## Subplots and Legends

```python
days = [1,2,3,4,5]
original_price = [20,34,25,40,58]
prediction = [18,30,25,41,55]
```

```python
#legends are label to lines plotted
plt.plot(days,original_price, 'g-o', label="original")
plt.plot(days,prediction, 'r-o', label = "Predicted")
plt.legend()
plt.title("Plot with legends")
plt.xlabel("days")
plt.ylabel("prices")
```

```python
#subplots = divide plots
plt.subplot(3,1,1) #(rows, column, index)
plt.plot(days,original_price, 'g-o', label="original")
plt.title("original prices over time")
plt.xlabel("days")
plt.ylabel("prices")
plt.legend()

plt.subplot(3,1,2) #(rows, column, index)
plt.plot(days,original_price, 'r-o', label="predicted")
plt.title("predicted prices over time")
plt.xlabel("days")
plt.ylabel("prices")
plt.legend()

#difference between original and predicted
#using list comprehension
price_diff = [op - pp for op,pp in zip(original_price,prediction)]
plt.subplot(3,1,3) #(rows, column, index)
plt.bar(days,price_diff, color='purple', label="price diff")
plt.title("price diff")
plt.xlabel("days")
plt.ylabel("price diff")
plt.legend()

plt.tight_layout()
```

### saving fig

```python
plt.plot(days,original_price, 'g-o', label="original")
plt.plot(days,prediction, 'r-o', label = "Predicted")
plt.legend()
plt.title("Plot with legends")
plt.xlabel("days")
plt.ylabel("prices")
plt.savefig("example.png",dpi=300)
```
