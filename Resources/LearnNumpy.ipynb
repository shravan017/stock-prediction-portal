## Appending and Inserting

```python
zeroArr
```

```python
zeros = np.zeros((3,2,3))
zeros
```

```python
ones = np.ones((3,2,3))
```

```python
ones
```

```python
# appending
new_arr = np.append(zeros,ones, axis = 0)
```

```python
new_arr
```

```python
#insert array
insert_array = np.insert(zeros, 2, ones, axis = 0)
insert_array
```

```python
#DELETE
arr4 = np.random.rand(3,2,3)
arr4
```

```python
arr5 = np.delete(arr4, 1, axis=1) #(arry, index, axis)
arr5
```

```python
# saving the array
np.save("save_array",arr5)
```

```python
# to load the file or array
np.load("save_array.npy")
```
